```python
types_of_movies_name=data["type"].unique()
numbers_types_of_movies_name=data['type'].value_counts().values
plt.pie(numbers_types_of_movies_name,labels=types_of_movies_name,autopct='%1.1f%%')
plt.title("Movies Vs Tv Show")
plt.figure(figsize=(15,6))
plt.show()
```

```python
print("All unique director->",data["director"].unique())
print("Total Number of unique directors->",data["director"].value_counts().count())
```

```python
name_of_max_5_directors=data['director'].value_counts().index[0:5]
print(name_of_max_5_directors)
plt.figure(figsize=(15,6))

movies_of_max_5_directors=data['director'].value_counts().values[0:5]
print(movies_of_max_5_directors)

plt.xlabel("Director")
plt.ylabel('Number of movies')
plt.title("Maximum Movie Director")
plt.bar(name_of_max_5_directors,movies_of_max_5_directors,color=["red",'blue','green','yellow','pink'])
```

```python
plt.figure(figsize=(15,6))
name_of_min_5_directors=data['director'].value_counts().index[-6:-1]
print(name_of_min_5_directors)

movies_of_min_5_directors=data['director'].value_counts().values[-6:-1]
print(movies_of_min_5_directors)

plt.title("Minimum Movies Director")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.bar(name_of_min_5_directors,movies_of_min_5_directors,color=['red','blue','green','yellow','pink'])
plt.show()
```
